```
install.packages("pacman")
library(pacman)
p_load(resampledata)
Chlorine <- Bangladesh$Chlorine
Chlorine <- Chlorine[!is.na(Chlorine)]
```

```
summary(Chlorine)
sd(Chlorine)

par(mfrow=c(1,2))
hist(Chlorine)
qqnorm(Chlorine)
qqline(Chlorine)
```

```
set.seed(9999)

n = length(Chlorine)
replicates = rep(0, 5000)
for (i in 1:5000){
  replicates[i] <- mean(sample(Chlorine, size=n, replace = TRUE ))
}

par(mfrow=c(1,2))
hist(replicates, xlab='replicates of sample mean', main='')
abline(v=mean(Chlorine), col='cyan')
qqnorm(replicates)
qqline(replicates)
```

```
as.numeric(round(quantile(replicates, c(0.025, 0.975)), 2))
```

```
# 95% CI = x_bar +/- z_c*sd/sqrt(n)
ci_lower <- mean(Chlorine) - 1.96 * sd(Chlorine)/sqrt(n)
ci_upper <- mean(Chlorine) + 1.96 * sd(Chlorine)/sqrt(n)

round(c(ci_lower, ci_upper), 2)
```

The bootstrap CI and the $Z_c$ CI are very close.  They also both have a width of about 50.
